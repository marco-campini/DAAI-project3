```python
import matplotlib.pyplot as plt

loss_list = []
acc_list = []
miou_list = []
files = ['GTA5_lr001_bs4_25epoch.rtf', 'GTA5_lr001_bs4_50epoch.rtf', 'GTA5_lr001_bs4_70epoch.rtf']
iterations = range(1, 71, 5)

for i, my_file in enumerate(files):
  with open(my_file, 'r') as f:
    for line in f:
      if line.startswith('loss for train') and not (len(loss_list) >= 25*(i+1) or len(loss_list) >= 70):
        loss_list.append(float(line.split(':')[-1].split('\\')[0]))
      if line.startswith('precision per pixel'):
        acc_list.append(float(line.split(':')[-1].split('\\')[0]))
      if line.startswith('mIoU for validation'):
        miou_list.append(float(line.split(':')[-1].split('\\')[0]))

plt.title("Loss")
plt.plot(loss_list, color="red")

plt.show()

plt.title("Accuracy")
plt.plot(iterations, acc_list, color="blue", label="acc")

plt.show()

plt.title("mIoU")
plt.plot(iterations, miou_list, color="green", label="mIoU")

# plt.legend()
plt.show()
```
